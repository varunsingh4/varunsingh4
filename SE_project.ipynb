```python
from bs4 import BeautifulSoup
import requests

url = "https://incubator.apache.org/projects/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")  # Correct parser

print(soup.prettify())
```
